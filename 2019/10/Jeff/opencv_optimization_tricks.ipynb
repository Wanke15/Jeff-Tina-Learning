```python
import cv2
import matplotlib.pyplot as plt
```

```python
%%timeit

img = cv2.UMat(cv2.imread("tree.jpg", cv2.IMREAD_COLOR))
imgUMat = cv2.UMat(img)
gray = cv2.cvtColor(imgUMat, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (7, 7), 1.5)
gray1 = cv2.Canny(gray, 0, 50)
# plt.imshow(gray1)
```

```python
img = cv2.UMat(cv2.imread("tree.jpg", cv2.IMREAD_COLOR))
imgUMat = cv2.UMat(img)
gray = cv2.cvtColor(imgUMat, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (7, 7), 1.5)
gray = cv2.Canny(gray, 0, 50)
gray1 = cv2.UMat.get(gray)
plt.imshow(gray1)
```
